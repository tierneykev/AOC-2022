```python


def calculate_file_sizes():    
    cwd = "/" #Current Working Dir
    fileSizes = {'/' : 0}

    for line in puzzleData:
        # ls command - do nothing    
        if line =='$ ls':
            continue
        # Dir listed in ls command - init fileSizes
        elif line.startswith('dir'):
            child_dir = cwd  + line[4:] + '/'
            fileSizes[child_dir] = 0
        # CD commands
        elif line.startswith('$ cd'):
            # Change Back to Root
            if line == '$ cd /':
                cwd = "/"
            # Change to parent
            elif line == '$ cd ..':
                cwd =  "/".join(cwd.split("/")[:-2]) + '/'
            # Change to Child ($ cd dirName)
            else:
                cwd += line.split(" ")[-1] + '/'

        # file result - add size to totals
        else:
            # 16394 tllvcdr.sjl
            fileSize = int(line.split(" ")[0])
            # Convert List to String to use as key

            fileSizes[cwd] += fileSize
    return fileSizes

def total_directory_sizes():
    # Loop over directory sizes - build out cumulative sizes
    cumTotals = {}
    for parent in fileSizes:
        for child in fileSizes:
            if child.startswith(parent):
                if parent in cumTotals.keys():
                    cumTotals[parent] += fileSizes[child]
                else:
                    cumTotals[parent] = fileSizes[child]    
    return cumTotals

def part_1():
    #Find all of the directories with a total size of at most 100000
    #What is the sum of the total sizes of those directories?
    fileSizeMax = 100000
    fileSizeMaxSum = 0
    for folder in cumTotals:
        if cumTotals[folder] <= fileSizeMax:
            #print(folder,cumTotals[folder])
            fileSizeMaxSum += cumTotals[folder]    
    return fileSizeMaxSum

def part_2():
    toalDiskSpace = 70000000
    updateFreeSpaceReq = 30000000
    
    totalUsedSpace = cumTotals['/']
    currentFreeSpace = toalDiskSpace - totalUsedSpace
    
    additionalFreeSpace = updateFreeSpaceReq - currentFreeSpace
    
    smallestDir = 0 #Smallest Dir That frees enough space
    for dir in cumTotals:
        # size of dir > needed additional free space
        if cumTotals[dir] > additionalFreeSpace:
            # size is smaller then current small match
            if cumTotals[dir] < smallestDir or smallestDir == 0:
                smallestDir = cumTotals[dir]
    
    return smallestDir
    

#inputFile = 'sample_input.txt'
inputFile = 'puzzle_input.txt'

with open(inputFile) as f:
    puzzleData = f.read().splitlines()

#Calculate Individual Directory Sizes    
fileSizes = calculate_file_sizes()
#Calculate Cumilative Directory Sizes
cumTotals = total_directory_sizes()    

#PART1 - Sum of cum directory sizes under limit
part1Solution = part_1()
print('Part 1',part1Solution)

part2Solution = part_2()
print('Part 2',part2Solution)
```
