```python
file = open("puzzle_input.txt")
data = file.readlines()

#Part 1
#The first column is 
#what your opponent is going to play: 
#A for Rock
#B for Paper
#C for Scissors

#The second column 
#what you should play in response: 
#X for Rock, 
#Y for Paper 
#Z for Scissors


xyz_to_abc = {
    'A' : 'X',
    'B' : 'Y',
    'C' : 'Z'
}

#Point Values Regardless of Results
#1 for Rock
#2 for Paper 
#3 for Scissors
play_values = {
    'X' : 1,
    'Y' : 2,
    'Z' : 3 
}

# Win/Lose/Draw Points
#0 if you lost 
#3 if the round was a draw 
#6 if you won

total = 0
for line in data:
    p1 = xyz_to_abc[line[0]]
    p2 = line[2]
    
    play_score = play_values[p2]
    # Draw
    if p1 == p2:
        play_score += 3
    # Rock beats scissors
    if p2 == 'X' and p1 == "Z":
        play_score += 6
    # Paper beats rock
    if p2 == 'Y' and p1 == "X":
        play_score += 6
    # scizzoe beats paper
    if p2 == 'Z' and p1 == "Y":
        play_score += 6
    total += play_score
    
    #print(p1,p2,play_score)
    
print(total)

# Part 2 - New Instructions
#X means you need to lose, 
#Y a draw 
#Z means you need to win

map = {
    # need to lose
    'X' :
    {
        'A' : 'C',
        'B' : 'A',
        'C' : 'B'
    },
    # need to win
    'Z' : {
        'A' : 'B',
        'B' : 'C',
        'C' : 'A'
    }
    
}
total2 = 0
for line in data:
    p1 = line[0]
    wld = line[2]
    
    #lose
    if wld == 'X':
        total2 += 0
    # Draw
    elif wld == 'Y':
        total2 += 3
        
    # Win
    else:
        total2 += 6
    
    # What to play
    if wld == 'Y':
        p2 = p1
    else:
        p2 = map[wld][p1]
    total2 += play_values[xyz_to_abc[p2]]
print(total2)
```
