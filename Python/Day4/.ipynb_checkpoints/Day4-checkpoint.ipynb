```python
import re
#file_name = 'sample_input.txt'
file_name = 'puzzle_input.txt'
with open(file_name) as f:
    lines = f.read().splitlines()

overlaps = 0
overlap_list = []
for line in lines:
    #print(line)
    match = re.match("(\d+)\-(\d+)\,(\d+)\-(\d+)",line)
    #1 inside #2 
    if match[1] >= match[3] and match[1] <= match[4]:
        if match[2] >= match[3] and match[2] <= match[4]:
            overlaps += 1
            overlap_list.append([match[1],match[2],match[3],match[4]])
            print('*',match[1],match[2],match[3],match[4])
    #2 inside #1
    elif match[3] >= match[1] and match[3] <= match[2]:
        if match[4] >= match[1] and match[4] <= match[2]:
            overlaps += 1
            
            print('+',match[1],match[2],match[3],match[4])   
       
print(overlaps)
    
```

```python
import re
#file_name = 'sample_input.txt'
file_name = 'puzzle_input.txt'
with open(file_name) as f:
    lines = f.read().splitlines()

overlaps = 0
overlap_list = []
for line in lines:
    elf1,elf2 = line.split(',')
    elf1start,elf1end = elf1.split('-')
    elf2start,elf2end = elf2.split('-')
    
    if int(elf1start) >= int(elf2start) and int(elf1end) <= int(elf2end):
        overlaps +=1
        print('*',line)
    elif int(elf2start) >= int(elf1start) and int(elf2end) <= int(elf1end):
        overlaps +=1
        print('+',line)
print(overlaps)
```
